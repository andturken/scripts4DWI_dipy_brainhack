```python
%matplotlib inline
from plot_surf_stat_map_3 import create_fig
from vtk_rw import read_vtk
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = '/Users/ghfc/Desktop/'
filename = 'P32_F16_parsed.vtk'
```

```python
vertex_array, face_array, data_array = read_vtk(path+filename)
```

```python
# random labels
data_ = np.array(np.random.rand(len(vertex_array)), dtype=np.float64)
```

```python
sns.set_context('poster')
sns.set_style("white")

create_fig(data= data_ ,coords=vertex_array, faces=face_array, azims=[0, 180], elevs=[180, 180],cmap='seismic', cbar=True)
plt.savefig(path+'P32_F16_fig.png')
```
