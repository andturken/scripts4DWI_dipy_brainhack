```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'E:/Celine/Ferret_b2n/'
subj1 = 'F25_P2'
subj2 = 'F10_P8'
subj3 = 'F21_P16'
subj4 = 'F16_P32'
subj5 = 'F01_Adult'
nameMat = '_grid0_444'
```

```python
#read conMat npz file

def importnpz (fname):
    conMatf = np.load(fname+'.npz')
    conMat1 = conMatf['arr_0']
    conMat1.shape
    return conMat1
```

```python
conMatF25 = importnpz(path+'/'+subj1+'/'+subj1+nameMat)
conMatF10 = importnpz(path+'/'+subj2+'/'+subj2+nameMat)
conMatF21 = importnpz(path+'/'+subj3+'/'+subj3+nameMat)
conMatF16 = importnpz(path+'/'+subj4+'/'+subj4+nameMat)
conMatF01 = importnpz(path+'/'+subj5+'/'+subj5+nameMat)
```

```python
diff2_8 = np.subtract(conMatF10,conMatF25)
#diff1 = np.subtract(conMatF25,conMatF10)
diff8_16 = np.subtract(conMatF21,conMatF10)
diff16_32 = np.subtract(conMatF16,conMatF21)
diff32_64 = np.subtract(conMatF01,conMatF16)
```

```python
plt.matshow(np.log10(diff2_8),interpolation='nearest',cmap=plt.cm.Blues)
plt.colorbar()
#plt.savefig(nameMat+'_log10.png')
plt.show()
```

```python
plt.matshow(np.log10(diff8_16),interpolation='nearest',cmap=plt.cm.Blues)
plt.colorbar()
#plt.savefig(nameMat+'_log10.png')
plt.show()
```

```python
plt.matshow(np.log10(diff16_32),interpolation='nearest',cmap=plt.cm.Blues)
plt.colorbar()
#plt.savefig(nameMat+'_log10.png')
plt.show()
```

```python
plt.matshow(np.log10(diff32_64),interpolation='nearest',cmap=plt.cm.Blues)
plt.colorbar()
#plt.savefig(nameMat+'_log10.png')
plt.show()
```
